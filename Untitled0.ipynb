```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('D3.csv')
X1 = data['X1'].values
X2 = data['X2'].values
X3 = data['X3'].values
Y = data['Y'].values

# Gradient Descent Function
def gradient_descent(X, Y, learning_rate, num_iterations):
    m = len(Y)
    theta0, theta1 = 0, 0
    losses = []

    for _ in range(num_iterations):
        h = theta0 + theta1 * X
        loss = (1/(2*m)) * np.sum((h - Y)**2)
        losses.append(loss)

        d_theta0 = (1/m) * np.sum(h - Y)
        d_theta1 = (1/m) * np.sum((h - Y) * X)

        theta0 -= learning_rate * d_theta0
        theta1 -= learning_rate * d_theta1

    return theta0, theta1, losses

# Hyperparameters
learning_rates = [0.1, 0.01, 0.001]
num_iterations = 1000
variables = {'X1': X1, 'X2': X2, 'X3': X3}

# Initialize a dictionary to store final losses for each variable and learning rate
final_losses = {var: {} for var in variables.keys()}

# Train for each variable with different learning rates and plot losses
for var_name, X in variables.items():
    plt.figure(figsize=(15, 5))
    plt.suptitle(f'Loss over Iterations for {var_name} at Different Learning Rates', fontsize=16)

    for i, lr in enumerate(learning_rates):
        theta0, theta1, losses = gradient_descent(X, Y, lr, num_iterations)
        final_loss = losses[-1]
        final_losses[var_name][lr] = final_loss

        # Plot loss over iterations
        plt.subplot(1, 3, i + 1)
        plt.plot(losses, color='blue')
        plt.title(f'Learning Rate: {lr}')
        plt.xlabel('Iterations')
        plt.ylabel('Loss')
        plt.grid(True)

    plt.tight_layout()
    plt.show()

# Report the final losses for each variable and learning rate
print("Final losses for each explanatory variable at different learning rates:")
for var_name, losses in final_losses.items():
    print(f"\n{var_name}:")
    for lr, loss in losses.items():
        print(f"Learning rate: {lr}, Final loss: {loss:.4f}")

# Determine which explanatory variable has the lowest loss for each learning rate
for lr in learning_rates:
    min_loss = float('inf')
    best_variable = None

    for var_name, losses in final_losses.items():
        if losses[lr] < min_loss:
            min_loss = losses[lr]
            best_variable = var_name

    print(f"\nFor learning rate {lr}, {best_variable} has the lowest loss: {min_loss:.4f}")
```
